# Program 1
Develop a program to create histograms for all numerical features and
analyze the distribution of each feature. Generate box plots for all numerical features
and identify any outliers. Use California Housing dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_california_housing

housing = fetch_california_housing(as_frame=True)
df = housing.frame

numerical_cols = df.select_dtypes(include=['number']).columns

print("Generating Histograms...")
for col in numerical_cols:
    plt.figure(figsize=(8, 6))
    sns.histplot(df[col], kde=True)
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()

print("\nGenerating Box Plots and identifying outliers...")
for col in numerical_cols:
    plt.figure(figsize=(8, 6))
    sns.boxplot(y=df[col])
    plt.title(f'Box Plot of {col}')
    plt.ylabel(col)
    plt.show()

    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)][col]
    if not outliers.empty:
        print(f"Outliers in {col}:")
        print(outliers)
    else:
        print(f"No significant outliers detected in {col}.")
```

# Program 2
Develop a program to Compute the correlation matrix to understand the
relationships between pairs of features. Visualize the correlation matrix using a
heatmap to know which variables have strong positive/negative correlations. Create a
pair plot to visualize pairwise relationships between features. Use California Housing
dataset.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_california_housing

housing = fetch_california_housing(as_frame=True)
df = housing.frame

numerical_cols = df.select_dtypes(include=['number']).columns

print("Generating Box Plots and identifying outliers...")
for col in numerical_cols:
    plt.figure(figsize=(8, 6))
    sns.boxplot(y=df[col])
    plt.title(f'Box Plot of {col}')
    plt.ylabel(col)
    plt.show()

    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)][col]
    if not outliers.empty:
        print(f"Outliers in {col}:")
        print(outliers)
    else:
        print(f"No significant outliers detected in {col}.")

print("\nComputing Correlation Matrix...")
correlation_matrix = df.corr()
print(correlation_matrix)

print("\nGenerating Correlation Heatmap...")
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')
plt.show()

print("\nGenerating Pair Plot...")
sns.pairplot(df[numerical_cols])
plt.suptitle('Pair Plot of Numerical Features', y=1.02)
plt.show()
```

# Program 3
Develop a program to implement Principal Component Analysis (PCA) for
reducing the dimensionality of the Iris dataset from 4 features to 2.


```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

iris = load_iris(as_frame=True)
X = iris.data
y = iris.target

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

pca_df = pd.DataFrame(data=X_pca, columns=['Principal Component 1', 'Principal Component 2'])
pca_df['target'] = y

print("Original Data Head:")
print(X.head())
print("\nPCA Transformed Data Head (2 Components):")
print(pca_df.head())

plt.figure(figsize=(8, 6))
sns.scatterplot(x='Principal Component 1', y='Principal Component 2', hue='target', data=pca_df, palette='viridis')
plt.title('2 Component PCA of Iris Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)
plt.show()
```

# Program 5
Develop a program to implement k-Nearest Neighbour algorithm to classify
the randomly generated 100 values of x in the range of [0,1]. Perform the following
based on dataset generated.
1. Label the first 50 points {x1,,x50} as follows: if (xi  0.5), then xi  Class1, else xi 
Class1
2. Classify the remaining points, x51,,x100 using KNN. Perform this for
k=1,2,3,4,5,20,30

```python
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)
x_values = np.random.rand(100)
df = pd.DataFrame({'x': x_values})

labels = np.array([''] * 100, dtype=object)
for i in range(50):
    labels[i] = 'Class1' if df['x'][i] <= 0.5 else 'Class2'
df['label'] = labels

X_train = df['x'].iloc[:50].values.reshape(-1, 1)
y_train = df['label'].iloc[:50].values
X_test = df['x'].iloc[50:].values.reshape(-1, 1)

k_values = [1, 2, 3, 4, 5, 20, 30]

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)

    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=X_train.flatten(), y=[0]*len(X_train), hue=y_train, palette={'Class1':'blue', 'Class2':'red'}, marker='o', s=100)
    sns.scatterplot(x=X_test.flatten(), y=[0]*len(X_test), hue=y_pred, palette={'Class1':'cyan', 'Class2':'magenta'}, marker='x', s=100)
    plt.title(f'KNN Classification for k={k}')
    plt.xlabel('x value')
    plt.yticks([])
    plt.legend()
    plt.grid(True)
    plt.show()
```

# Program 6
Implement the non-parametric Locally Weighted Regression algorithm in
order to fit data points. Select appropriate data set for your experiment and draw graphs

```python
import numpy as np
import matplotlib.pyplot as plt

def gaussian_kernel(x, x_i, tau):
    return np.exp(-((x - x_i)**2) / (2 * tau**2))

def locally_weighted_regression(X, y, x_query, tau):
    m = X.shape[0]
    y_pred = np.zeros(len(x_query))
    X_bias = np.c_[np.ones(m), X]  # Add bias once outside loop

    for i, x_val in enumerate(x_query):
        W = np.diag([gaussian_kernel(x_val, X[j], tau) for j in range(m)])
        try:
            theta = np.linalg.inv(X_bias.T @ W @ X_bias) @ X_bias.T @ W @ y
            y_pred[i] = np.array([1, x_val]) @ theta
        except np.linalg.LinAlgError:
            y_pred[i] = np.nan
    return y_pred

np.random.seed(0)
X = np.random.rand(100) * 10
y = np.sin(X) + np.random.normal(0, 0.5, 100)

plt.figure(figsize=(10, 6))
plt.scatter(X, y, label='Original Data', alpha=0.7)

x_query = np.linspace(X.min(), X.max(), 200)  # 1D array here
tau_values = [0.5, 1.0, 2.0]

for tau in tau_values:
    y_pred = locally_weighted_regression(X, y, x_query, tau)
    plt.plot(x_query, y_pred, label=f'LWR (tau={tau})')

plt.title('Locally Weighted Regression with Different Tau Values')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

# Program 7
Develop a program to demonstrate the working of Linear Regression and
Polynomial Regression. Use Boston Housing Dataset for Linear Regression and Auto MPG
Dataset (for vehicle fuel efficiency prediction) for Polynomial Regression

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.impute import SimpleImputer

# Paths
boston_path = "/content/BostonHousing.csv"
auto_path = "/content/auto-mpg.csv"

boston = pd.read_csv(boston_path)
X_boston, y_boston = boston.iloc[:, :-1], boston['medv']
X_train, X_test, y_train, y_test = train_test_split(X_boston, y_boston, test_size=0.2, random_state=42)
lr = LinearRegression().fit(X_train, y_train)
y_pred = lr.predict(X_test)
print("Boston Linear Regression:\nMSE:", mean_squared_error(y_test, y_pred), "R:", r2_score(y_test, y_pred))
plt.scatter(y_test, y_pred, c='blue'); plt.plot([0, 50], [0, 50], 'r'); plt.xlabel("True"); plt.ylabel("Predicted")
plt.title("Boston Housing - Linear Regression"); plt.show()

auto = pd.read_csv(auto_path)
imp = SimpleImputer(strategy='mean')
auto[['horsepower', 'weight']] = imp.fit_transform(auto[['horsepower', 'weight']])
X_auto, y_auto = auto[['horsepower', 'weight']], auto['mpg']
X_train, X_test, y_train, y_test = train_test_split(X_auto, y_auto, test_size=0.2, random_state=42)
poly_X_train = PolynomialFeatures(2).fit_transform(X_train)
poly_X_test = PolynomialFeatures(2).fit(X_train).transform(X_test)
pr = LinearRegression().fit(poly_X_train, y_train)
y_pred = pr.predict(poly_X_test)
print("Auto MPG Polynomial Regression:\nMSE:", mean_squared_error(y_test, y_pred), "R:", r2_score(y_test, y_pred))
plt.scatter(y_test, y_pred, c='green'); plt.plot([0, 50], [0, 50], 'r'); plt.xlabel("True"); plt.ylabel("Predicted")
plt.title("Auto MPG - Polynomial Regression"); plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt

data = pd.read_csv('/content/breastcancer_modified.csv')
X, y = data.drop(['diagnosis'], axis=1), data['diagnosis']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

print(f"Accuracy: {accuracy_score(y_test, y_pred)*100:.2f}%")
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))

plt.figure(figsize=(20,10))
plot_tree(clf, filled=True, feature_names=X.columns, class_names=clf.classes_.astype(str), rounded=True)
plt.show()

sample = pd.DataFrame({'radius_mean':[15.0],'texture_mean':[20.0],'perimeter_mean':[100.0],'area_mean':[500.0],
'smoothness_mean':[0.1],'compactness_mean':[0.2],'concavity_mean':[0.3],'concave points_mean':[0.15],
'symmetry_mean':[0.3],'fractal_dimension_mean':[0.05],'radius_se':[0.1],'texture_se':[0.2],'perimeter_se':[0.05],
'area_se':[200.0],'smoothness_se':[0.02],'compactness_se':[0.05],'concavity_se':[0.1],'concave points_se':[0.03],
'symmetry_se':[0.06],'fractal_dimension_se':[0.02],'radius_worst':[25.0],'texture_worst':[25.0],
'perimeter_worst':[150.0],'area_worst':[1200.0],'smoothness_worst':[0.1],'compactness_worst':[0.5],
'concavity_worst':[0.4],'concave points_worst':[0.3],'symmetry_worst':[0.5],'fractal_dimension_worst':[0.1]})

sample_scaled = scaler.transform(sample)
pred = clf.predict(sample_scaled)
print("New Sample Prediction:", pred[0])
```

```python
import scipy.io
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

data = scipy.io.loadmat('/content/olivettifaces.mat')
X = data['faces'].T
y = np.array([i // 10 for i in range(X.shape[0])])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
model = GaussianNB()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print(f"Naive Bayes Classifier Accuracy: {accuracy_score(y_test, y_pred)*100:.2f}%")
print(classification_report(y_test, y_pred, zero_division=0))
print(confusion_matrix(y_test, y_pred))

fig, axes = plt.subplots(2, 5, figsize=(10, 5))
for i, ax in enumerate(axes.ravel()):
    ax.imshow(X_test[i].reshape(64, 64), cmap='gray')
    ax.set_title(f"Pred: {y_pred[i]}\nActual: {y_test[i]}")
    ax.axis('off')
plt.tight_layout()
plt.show()
```

# Program 10

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

df = pd.read_csv('/content/breastcancer_modified.csv')
df['diagnosis_encoded'] = df['diagnosis'].map({'M': 1, 'B': 0})
df = df.drop(columns=[col for col in ['id', 'diagnosis'] if col in df.columns])
features = df.drop(columns=['diagnosis_encoded']).select_dtypes(include=[np.number])
features = features.dropna(axis=1, how='any')
X_scaled = StandardScaler().fit_transform(features)
kmeans = KMeans(n_clusters=2, random_state=42, n_init=10)
df['Cluster'] = kmeans.fit_predict(X_scaled)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
df['PCA1'], df['PCA2'] = X_pca[:, 0], X_pca[:, 1]
plt.figure(figsize=(8, 6))
sns.scatterplot(x='PCA1', y='PCA2', hue='Cluster', data=df, palette='viridis', s=80)
plt.scatter(*pca.transform(kmeans.cluster_centers_).T, c='red', marker='X', s=200, label='Centroids')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.title('K-Means Clustering Visualization')
plt.legend()
plt.show()
print("Cluster Centers:\n", kmeans.cluster_centers_)
comparison = pd.crosstab(df['diagnosis_encoded'], df['Cluster'], rownames=['Actual'], colnames=['Cluster'])
print("\nComparison of Clusters vs Actual Diagnosis:")
print(comparison)
```
